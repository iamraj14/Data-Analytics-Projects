```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "C:/Users/PRAJAKTA/Desktop/Financial Analytics data.csv"
data = pd.read_csv(file_path)
```

```python
print(data.head())
```

```python
data_cleaned = data.drop(columns=['Unnamed: 4'], errors='ignore')
```

```python
summary_stats = data_cleaned.describe()
print("Summary Statistics:\n", summary_stats)
```

```python
correlation_matrix = data_cleaned[['Mar Cap - Crore', 'Sales Qtr - Crore']].corr()
print("\nCorrelation Matrix:\n", correlation_matrix)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['Mar Cap - Crore'], bins=30, kde=True)
plt.title('Distribution of Market Capitalization')
plt.xlabel('Market Capitalization (in Crores)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['Sales Qtr - Crore'], bins=30, kde=True, color='orange')
plt.title('Distribution of Quarterly Sales')
plt.xlabel('Quarterly Sales (in Crores)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Mar Cap - Crore', y='Sales Qtr - Crore', data=data_cleaned)
plt.title('Market Capitalization vs Quarterly Sales')
plt.xlabel('Market Capitalization (in Crores)')
plt.ylabel('Quarterly Sales (in Crores)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
avg_market_cap = data_cleaned['Mar Cap - Crore'].mean()
print(f"Average Market Capitalization: {avg_market_cap:.2f} Crores")
```

```python
median_market_cap = data_cleaned['Mar Cap - Crore'].median()
print(f"Median Market Capitalization: {median_market_cap:.2f} Crores")
```

```python
avg_quarterly_sales = data_cleaned['Sales Qtr - Crore'].mean()
print(f"Average Quarterly Sales: {avg_quarterly_sales:.2f} Crores")
```

```python
median_quarterly_sales = data_cleaned['Sales Qtr - Crore'].median()
print(f"Median Quarterly Sales: {median_quarterly_sales:.2f} Crores")
```

```python
market_cap_range = data_cleaned['Mar Cap - Crore'].max() - data_cleaned['Mar Cap - Crore'].min()
print(f"Market Capitalization Range: {market_cap_range:.2f} Crores")
```

```python
quarterly_sales_range = data_cleaned['Sales Qtr - Crore'].max() - data_cleaned['Sales Qtr - Crore'].min()
print(f"Quarterly Sales Range: {quarterly_sales_range:.2f} Crores")
```
